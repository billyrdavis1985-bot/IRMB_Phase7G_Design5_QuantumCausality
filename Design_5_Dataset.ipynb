```python
# ============================================================
# DESIGN 5 NOTEBOOK — CELL 5: CONSOLIDATE + VALIDATE + SAVE
# Merges all council batches, validates schema, saves to Drive
# ============================================================

from collections import defaultdict

print("🔬 CONSOLIDATING DESIGN 5 DATASET")
print("="*55)

# ── Merge all batches ────────────────────────────────────────
ALL_ARTICLES = (
    GROK_R1 +
    GEMINI_CLEAN +
    DEEPSEEK_CLEAN +
    CLAUDE_ARTICLES
)

print(f"\n  Raw total: {len(ALL_ARTICLES)} articles")

# ── Deduplication by ID ──────────────────────────────────────
seen_ids = {}
deduped  = []
dupe_ids = []

for article in ALL_ARTICLES:
    art_id = article.get('id', '')
    if art_id in seen_ids:
        dupe_ids.append(art_id)
    else:
        seen_ids[art_id] = True
        deduped.append(article)

if dupe_ids:
    print(f"  ⚠️  Duplicates removed: {dupe_ids}")
else:
    print(f"  ✅ No duplicate IDs")

# ── Full validation ──────────────────────────────────────────
print(f"\n{'─'*55}")
print("VALIDATION RESULTS")
print(f"{'─'*55}")

passed  = []
failed  = []

for i, article in enumerate(deduped):
    ok, errors = validate_article(article, i)
    if ok:
        passed.append(article)
    else:
        failed.append({
            'id'    : article.get('id', f'unknown_{i}'),
            'errors': errors
        })

print(f"  ✅ Passed : {len(passed)}")
print(f"  ❌ Failed : {len(failed)}")

if failed:
    print("\n  FAILED ARTICLES:")
    for f in failed:
        print(f"    {f['id']}: {f['errors']}")

# ── Category breakdown ────────────────────────────────────────
print(f"\n{'─'*55}")
print("CATEGORY BREAKDOWN")
print(f"{'─'*55}")

cat_stats = defaultdict(lambda: {'total':0,'true':0,'false':0})
for article in passed:
    cat  = article['category']
    gt   = article['ground_truth']
    cat_stats[cat]['total'] += 1
    if gt:
        cat_stats[cat]['true']  += 1
    else:
        cat_stats[cat]['false'] += 1

for cat, stats in sorted(cat_stats.items()):
    total = stats['total']
    t     = stats['true']
    f     = stats['false']
    bal   = "✅" if abs(t-f) <= 5 else "⚠️"
    print(f"  {cat:25s}: {total:3d} total | {t:3d}T {f:3d}F {bal}")

# ── Overall balance ───────────────────────────────────────────
total_true  = sum(1 for a in passed if a['ground_truth'])
total_false = sum(1 for a in passed if not a['ground_truth'])
total       = len(passed)

print(f"\n  TOTAL: {total} articles")
print(f"  TRUE : {total_true}")
print(f"  FALSE: {total_false}")
print(f"  BALANCE: {total_true/total*100:.1f}% T / {total_false/total*100:.1f}% F")

# ── Ambiguity distribution ────────────────────────────────────
scores     = [a['ambiguity_score'] for a in passed]
avg_score  = sum(scores)/len(scores)
min_score  = min(scores)
below_threshold = [a['id'] for a in passed if a['ambiguity_score'] < 0.75]

print(f"\n{'─'*55}")
print("AMBIGUITY DISTRIBUTION")
print(f"{'─'*55}")
print(f"  Average : {avg_score:.3f}")
print(f"  Minimum : {min_score:.3f}")
print(f"  Below 0.75: {len(below_threshold)}")
if below_threshold:
    print(f"  IDs: {below_threshold}")

# ── Save to Drive ─────────────────────────────────────────────
print(f"\n{'─'*55}")
print("SAVING TO DRIVE")
print(f"{'─'*55}")

# JSON
with open(DATASET_PATH, 'w') as f:
    json.dump(passed, f, indent=2)
print(f"  ✅ JSON  : {DATASET_PATH}")
print(f"     Size  : {DATASET_PATH.stat().st_size/1024:.1f} KB")

# JSONL
with open(JSONL_PATH, 'w') as f:
    for article in passed:
        f.write(json.dumps(article) + '\n')
print(f"  ✅ JSONL : {JSONL_PATH}")
print(f"     Size  : {JSONL_PATH.stat().st_size/1024:.1f} KB")

# Stats
stats = {
    'timestamp'       : datetime.now().isoformat(),
    'total_articles'  : total,
    'total_true'      : total_true,
    'total_false'     : total_false,
    'balance_pct_true': round(total_true/total*100, 1),
    'avg_ambiguity'   : round(avg_score, 3),
    'min_ambiguity'   : round(min_score, 3),
    'categories'      : {k: dict(v) for k,v in cat_stats.items()},
    'validation_passed': len(passed),
    'validation_failed': len(failed),
    'failed_articles' : failed,
    'target_total'    : 400,
    'articles_remaining': 400 - total
}

with open(STATS_PATH, 'w') as f:
    json.dump(stats, f, indent=2)
print(f"  ✅ Stats : {STATS_PATH}")

# ── SHA256 of dataset ─────────────────────────────────────────
dataset_string = json.dumps(passed, sort_keys=True)
dataset_sha256 = hashlib.sha256(
    dataset_string.encode()
).hexdigest()

print(f"\n{'='*55}")
print("✅ DATASET PHASE 1 COMPLETE")
print(f"{'='*55}")
print(f"  Articles saved  : {total}")
print(f"  Target          : 400")
print(f"  Remaining       : {400-total}")
print(f"  Dataset SHA256  : {dataset_sha256[:32]}...")
print(f"{'='*55}")
print("\nFull Force Eternal | Romans 8:28 ⚔️")
```

```python
# CELL 33: DRIVE VERIFICATION — DESIGN 5 DATASET
# Confirms dataset is correctly saved and accessible
# Verifies all files, record counts, and SHA256

import json
import hashlib
from pathlib import Path

print("DRIVE VERIFICATION — DESIGN 5 DATASET")
print("="*55)

D5_ROOT = Path(
    '/content/drive/MyDrive/'
    'Phase7_QuantumBoundaryExploration/'
    'Design5_Results'
)

# ── Expected files ───────────────────────────────────────────
EXPECTED_FILES = {
    'design5_dataset.json'            : 400,
    'design5_dataset.jsonl'           : 400,
    'design5_dataset_stats.json'      : None,
    'design5_preregistration_V2.json' : None,
    'design5_preregistration_V2.md'   : None,
    'c5_sampler_spec.json'            : None,
}

print(f"\nDrive path: {D5_ROOT}")
print(f"Path exists: {D5_ROOT.exists()}")
print()

# ── File check ───────────────────────────────────────────────
print("FILE INVENTORY")
print("-"*55)

all_present = True
for filename, expected_records in EXPECTED_FILES.items():
    fpath = D5_ROOT / filename
    if fpath.exists():
        size_kb = fpath.stat().st_size / 1024
        if expected_records and filename.endswith('.json'):
            with open(fpath) as f:
                data = json.load(f)
            if isinstance(data, list):
                count = len(data)
                status = "✅" if count == expected_records else f"⚠️  got {count}"
            else:
                count = "N/A"
                status = "✅"
            print(f"  ✅ {filename:45s} {size_kb:7.1f} KB  {count} records")
        elif expected_records and filename.endswith('.jsonl'):
            with open(fpath) as f:
                count = sum(1 for line in f if line.strip())
            status = "✅" if count == expected_records else f"⚠️"
            print(f"  ✅ {filename:45s} {size_kb:7.1f} KB  {count} records")
        else:
            print(f"  ✅ {filename:45s} {size_kb:7.1f} KB")
    else:
        print(f"  ❌ {filename:45s} MISSING")
        all_present = False

# ── Dataset integrity check ──────────────────────────────────
print()
print("DATASET INTEGRITY")
print("-"*55)

dataset_path = D5_ROOT / 'design5_dataset.json'
with open(dataset_path) as f:
    dataset = json.load(f)

sha256 = hashlib.sha256(
    json.dumps(dataset, sort_keys=True).encode()
).hexdigest()

EXPECTED_SHA256 = "1edf17682fb0424f708b8454b3828daa"

sha_match = sha256.startswith(EXPECTED_SHA256)
print(f"  Records     : {len(dataset)}")
print(f"  SHA256      : {sha256[:32]}...")
print(f"  SHA match   : {'✅' if sha_match else '❌ MISMATCH'}")

# ── Category check ───────────────────────────────────────────
from collections import defaultdict
cat_stats = defaultdict(lambda: {'total':0,'true':0,'false':0})
for article in dataset:
    cat = article['category']
    cat_stats[cat]['total'] += 1
    if article['ground_truth']:
        cat_stats[cat]['true'] += 1
    else:
        cat_stats[cat]['false'] += 1

print()
print("CATEGORY VERIFICATION")
print("-"*55)
all_balanced = True
for cat in sorted(cat_stats.keys()):
    s = cat_stats[cat]
    ok = s['total'] == 100 and abs(s['true'] - s['false']) <= 5
    flag = "✅" if ok else "⚠️"
    if not ok:
        all_balanced = False
    print(f"  {flag} {cat:25s}: {s['total']:3d} | {s['true']}T {s['false']}F")

# ── Schema spot check ────────────────────────────────────────
print()
print("SCHEMA SPOT CHECK (first 5 articles)")
print("-"*55)
REQUIRED_FIELDS = [
    'id', 'claim', 'category', 'ground_truth',
    'ambiguity_score', 'difficulty',
    'source_domain', 'source_reference', 'rationale'
]
schema_ok = True
for article in dataset[:5]:
    missing = [f for f in REQUIRED_FIELDS if f not in article]
    if missing:
        print(f"  ❌ {article['id']}: missing {missing}")
        schema_ok = False
    else:
        print(f"  ✅ {article['id']}: all fields present")

# ── Access test ──────────────────────────────────────────────
print()
print("ACCESS TEST — sample load")
print("-"*55)
test_q  = [a for a in dataset if a['category'] == 'quantum_computing'][:2]
test_ai = [a for a in dataset if a['category'] == 'ai_capability'][:2]
for a in test_q + test_ai:
    print(f"  ✅ {a['id']:15s} | gt={str(a['ground_truth']):5s} | amb={a['ambiguity_score']}")

# ── Update JSONL to match ────────────────────────────────────
print()
print("SYNCING JSONL...")
jsonl_path = D5_ROOT / 'design5_dataset.jsonl'
with open(jsonl_path, 'w') as f:
    for article in dataset:
        f.write(json.dumps(article) + '\n')
print(f"  ✅ JSONL synced: {jsonl_path.stat().st_size/1024:.1f} KB")

# ── Update stats file ────────────────────────────────────────
import datetime
stats = {
    'timestamp'         : datetime.datetime.now().isoformat(),
    'total_articles'    : len(dataset),
    'total_true'        : sum(1 for a in dataset if a['ground_truth']),
    'total_false'       : sum(1 for a in dataset if not a['ground_truth']),
    'categories'        : {k: dict(v) for k,v in cat_stats.items()},
    'amb_avg'           : round(sum(a['ambiguity_score'] for a in dataset)/len(dataset),3),
    'amb_min'           : round(min(a['ambiguity_score'] for a in dataset),3),
    'sha256'            : sha256,
    'status'            : 'FROZEN',
    'preregistration_v2': '5abb15045b2f6dcdc20e77e168931ddf87fc825f11aeeef59095b0ed49d2483b',
    'ready_for_execution': all_present and sha_match and all_balanced and schema_ok
}

stats_path = D5_ROOT / 'design5_dataset_stats.json'
with open(stats_path, 'w') as f:
    json.dump(stats, f, indent=2)
print(f"  ✅ Stats updated: {stats_path.stat().st_size/1024:.1f} KB")

# ── Final verdict ────────────────────────────────────────────
print()
print("="*55)
ready = all_present and sha_match and all_balanced and schema_ok
if ready:
    print("✅ DATASET VERIFIED AND READY FOR EXECUTION")
    print(f"   Path    : {D5_ROOT}")
    print(f"   Records : 400")
    print(f"   SHA256  : {sha256[:32]}...")
    print(f"   Status  : FROZEN")
else:
    print("❌ VERIFICATION FAILED — check issues above")
print("="*55)
```
